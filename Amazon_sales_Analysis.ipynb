```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("C:\\Users\\ADMIN\\Documents\\E_commerce_project_dataset\\Amazon Sale Report.csv")
df
```

Checking missing data

```python
df.isnull().sum()
```

Filling missing data with Mode

```python
df['Courier Status'].fillna(df['Courier Status'].mode()[0],inplace= True)
```

```python
df['currency'].fillna(df['currency'].mode()[0],inplace= True)
```

Filling missing data with Median

```python
df['Amount'].fillna(df['Amount'].median(),inplace= True)
```

```python
df['ship-city'].fillna(df['ship-city'].mode()[0],inplace= True)
```

```python
df['ship-state'].fillna(df['ship-state'].mode()[0],inplace= True)
```

```python
df['ship-postal-code'].fillna(df['ship-postal-code'].mode()[0],inplace= True)
```

```python
df['ship-country'].fillna(df['ship-country'].mode()[0],inplace= True)
```

```python
df['promotion-ids'].fillna(df['promotion-ids'].mode()[0],inplace= True)
```

```python
df['fulfilled-by'].fillna(df['fulfilled-by'].mode()[0],inplace= True)
```

```python
df.describe()
```

Sales Performance Analysis

```python
 df['Sales']= df['Qty']*df['Amount']
```

```python
top_sold_category= df.groupby('Category')['Sales'].sum().sort_values(ascending= False)
top_sold_category
```

```python
top_sold_category.plot(kind= 'bar',cmap= 'Dark2')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

```python
df['Date']= pd.to_datetime(df['Date'])
df['Month']= df['Date'].dt.month
sales_by_month= df.groupby('Month')['Sales'].sum().sort_values(ascending= False)
sales_by_month
```

```python
sns.lineplot(x= 'Month',y= 'Sales',data= df,marker= 'o')
plt.title('Sales by Month')
plt.show()
```

```python
sales_by_states= df.groupby('ship-state')['Sales'].sum().sort_values(ascending= False).head(10)
sales_by_states
```

```python
sales_by_states.plot(kind= 'barh')
plt.xlabel('Sales')
plt.ylabel('State')
plt.title('Sales by State')
plt.show()
```

Customer Behavior Insights

```python
shipping_preferences= df.groupby('ship-service-level')['Order ID'].count()
shipping_preferences
```

```python
shipping_preferences.plot(kind= 'pie',autopct= '%1.2f%%',cmap= 'BrBG')
```

```python
B2B_orders_and_consumer_orders= df[df['B2B']==True]['Order ID'].count(),df[df['B2B']==False]['Order ID'].count()
B2B_orders_and_consumer_orders
```

```python
Courier_Status= df.groupby('Courier Status')['Order ID'].count()
Courier_Status
```

```python
Courier_Status.plot(kind= 'pie',autopct= '%1.2f%%')
plt.ylabel("")
plt.show()
```

Profitability Analysis

```python
total_sales= df['Amount'].sum()
total_orders= df['Order ID'].nunique()
aov= total_sales/total_orders
print(f"Average Order value: {aov}")
aov_by_category= df.groupby('Category')['Amount'].mean().sort_values(ascending= False)
aov_by_category
```

```python
aov_by_category.plot(kind= 'pie',autopct= '%1.2f%%')
```

 Product Performance

```python
sales_by_size= df.groupby('Size')[['Sales','Category']].sum().sort_values(by= 'Sales',ascending= False)
sales_by_size
```

```python
sales_by_size.plot(kind= 'barh')
plt.xlabel('Sales')
plt.ylabel('Size')
plt.title('Sales by Size')
plt.show()
```

Stock Keeping Unit (SKU) - Level Performance

```python
Sku_sales_and_orders= df.groupby(['SKU','Sales'])['Order ID'].count().sort_values(ascending = False).head(10)
Sku_sales_and_orders
```

```python
Sku_sales_and_orders.plot(kind= 'bar',color= 'g')
plt.xlabel('SKU Sales')
plt.ylabel('No.of Order Receive')
plt.title('Stock Level Unit Performance')
plt.show()
```
